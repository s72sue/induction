# Nengo Model

```python
import nengo
import nengo.spa as spa
from nengo.spa import Vocabulary
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim=5
vocab = Vocabulary(dimensions=dim, max_similarity=0.02) #, unitary=True)
#zero = spa.SemanticPointer(dim)

incr = vocab.parse("INCREMENT")
print (incr.v)
print (np.linalg.norm(incr.v))
incr.make_unitary()
print (incr.v)
print (np.linalg.norm(incr.v))
```

```python
d = 32
sp = spa.SemanticPointer(d)
sp.make_unitary()
lengths = []
for i in range(1, 10):
    lengths.append(np.linalg.norm(sp.v))
    s = spa.SemanticPointer(d)
    s.make_unitary()
    sp = sp*sp + s
    sp.normalize()
    
plt.plot(sp.v)
```
